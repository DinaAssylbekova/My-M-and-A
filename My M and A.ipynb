### Writing everything into Database   

```python
def my_m_and_a():
    connection = sql.connect('customers.db')                # to connect to database
    cursor = connection.cursor()                            # to execute SQL commands
 
    cursor.execute('DROP TABLE IF EXISTS plastic_free_boutique')

                       # '''Create table...''' to ignore spaces
    cursor.execute('''CREATE TABLE plastic_free_boutique
                     (gender TEXT,     
                      first_name TEXT, 
                      last_name TEXT, 
                      email TEXT, 
                      age TEXT, 
                      city TEXT,
                      country TEXT,
                      created_at TEXT,
                      referral TEXT)''')


    for index, row in df.iterrows():
        gender = row['gender']
        first_name = row['first_name']
        last_name = row['last_name']
        email = row['email']
        age = row['age']
        city = row['city']
        country = row['country']
        created_at = row['created_at']
        referral = row['referral']

        cursor.execute('INSERT INTO plastic_free_boutique VALUES (?,?,?,?,?,?,?,?,?)', (gender, first_name, last_name, email, age, city, country, created_at, referral))
        
    cursor.execute("SELECT * FROM plastic_free_boutique")
    print(cursor.fetchall())
```

```python
my_m_and_a()
```
